```python
# Fixed version - works better in Colab!
!pip install diffusers transformers accelerate torch

from diffusers import StableDiffusionPipeline
import torch

# Check if GPU is working
if torch.cuda.is_available():
    print("GPU is working! 🎉")
else:
    print("GPU not found - go back to Runtime menu and check GPU setting")

model_id = "nitrosocke/Ghibli-Diffusion"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# This line is changed to work better:
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# Third section - make your first Ghibli image
prompt = "a cute forest spirit from studio ghibli, smiling, magical atmosphere"
image = pipe(prompt).images[0]
image.save("ghibli_magic.png")
image
```
